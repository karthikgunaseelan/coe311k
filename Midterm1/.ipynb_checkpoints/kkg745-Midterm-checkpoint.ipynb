```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 5, 9, 13, 17, 19, 22, 25, 27, 28, 29, 33, 37, 41, 42, 43, 45, 49, 54, 56])
y = np.array([1.7, 0.1, 2.3, 2.7, 1.7, 5.0, 3.0, 1.5, 3.5, 1.8, 1.3, 2.5, 3.1, -5.1, -28.1, 
              33.8, 6.3, -1.6, 2.4, 3.3])

plt.scatter(x,y, label="Given Data Points")
plt.title("GDP Growth vs. Quarter")
plt.xlabel('Quarter')
plt.ylabel('GDP Growth (%)')
plt.legend()
plt.show()
```

This scatter plot visualizes the data set provided by the U.S. Bureau of Economic Analysis, analyzing quarterly U.S. GDP growth from 2010 to 2023. In the data set, 20 of the 56 quarterly data points are included. Therefore, we will be applying 3 methods to interpolate the rest of the data: cublic spline interpolation, polynomial fitting, and least squares linear regression.

PART A:

```python
def setup_tridiagonal_matrix(x, y):
    
    n = len(x)
    h = np.diff(x)

    A = np.zeros(n - 2)
    B = np.zeros(n - 2)
    C = np.zeros(n - 2)
    D = np.zeros(n - 2)

    for i in range(1, n - 1):
        A[i - 1] = h[i-1]
        B[i - 1] = 2*(h[i-1]+h[i])
        C[i - 1] = h[i]
        D[i - 1] = 6*((y[i+1]-y[i])/h[i] - (y[i]-y[i-1])/h[i-1])
        
    return A, B, C, D
```

```python
def thomas_algorithm(a, b, c, d):
    
    n = len(b)
   
    for i in range(1, n):
        w = a[i-1] / b[i-1]
        b[i] = b[i] - w * c[i-1]
        d[i] = d[i] - w * d[i-1]
  
    x = np.zeros(n)
    x[-1] = d[-1] / b[-1]
    
    for i in range(n-2, -1, -1):
        x[i] = (d[i] - c[i] * x[i+1]) / b[i]
        
    return x
```

```python
def evaluate_spline(x_data, y_data, M, x_query):

    n = len(x_data)
    h = np.diff(x_data) # step sizes
    i = np.searchsorted(x_data, x_query, side='right') - 1
    i = np.clip(i, 0, n - 2)
    hi = h[i]
    dx_left = x_query - x_data[i]
    dx_right = x_data[i+1] - x_query
    S = ((M[i] / (6*hi)) * dx_right**3 + (M[i+1] / (6*hi)) * dx_left**3 + 
         (y_data[i] /hi - M[i] *hi/6) * dx_right + (y_data[i+1] /hi - M[i+1] *hi/6) * dx_left)

    return S
```

```python
A, B, C, D = setup_tridiagonal_matrix(x, y)

M = np.zeros(len(x))
M[1:-1] = thomas_algorithm(A, B, C, D)

x_cspline = np.arange(x[0], x[-1], 0.01)
y_cspline = evaluate_spline(x, y, M, x_cspline)

plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x_cspline, y_cspline, color = 'red', label = 'Cubic Spline')
plt.xlabel('Quarter')
plt.ylabel('GDP Growth (%)')
plt.title('U.S Quarterly GDP Growth Rate from 2010 to 2023')
plt.legend()
plt.show()

# Condition Number
m = len(B)
X = np.zeros((m, m)) 
np.fill_diagonal(X, B)
np.fill_diagonal(X[1:], A)
np.fill_diagonal(X[:, 1:], C)
condition = np.linalg.cond(X)
print(f'Condition Number: {condition}')
```

1. To create the Cubic Spline Interpolation, we must set up a system of equations that relates the second derivatives at each interior knot through the Tridiagonal Matrix. For 20 data points, we solve for 18 unknowns (corresponding to the second derivatives at the interior knots) using the Thomas Algorithm, a simplified form of Gaussian Elimination for tridiagonal systems. These conditions ensure the continuity conditions of the spline curve, as the first and second derivative at the interior knots must be continuous. For a spline curve, the second derivatives at the endpoints are set equal to 0, ensuring boundary conditions.

```python
x_quarter_interp = np.arange(x[0], x[-1]+1, 1)
y_quarter_interp = evaluate_spline(x, y, M, x_quarter_interp)

plt.scatter(x_quarter_interp, y_quarter_interp, color='b', label="Quarterly Cubic Spline Values")
plt.plot(x_cspline, y_cspline, 'r-', label="Cubic Spline")
plt.title('U.S Quarterly GDP Growth Rate from 2010 to 2023')
plt.xlabel('Quarter')
plt.ylabel('GDP Growth (%)')
plt.legend()
plt.show()
```

2. The interpolant appears visually smooth for the most part, gradually transitioning between points in the data set. However, its behavior becomes questionable near outliers, such as 2020 Q2 and 2020 Q3. Since the cubic spline curve must pass through all data points, it is very sensitive to noisy data, causing noticeable oscillations in nearby areas. As a result, the spline accurately reproduces the given data, but it may overfit outliers and produce unrealistic behaviors between data points, making its reasonability questionable.

3. The Runge Phenomenon describes oscillatory behavior that can appear when the model overfits the data and interpolation is forced. Since the cubic spline curve must pass through every given data point, including outliers in 2020 Q2 and 2020 Q3, numerical stability is traded off for exact interpolation, which may make the curve less reasonable between certain points. As a result, smoothing splines or weighted least squares would likely be preferable if the the goal is to estimate the overall GDP growth trend, as it reduces the influence of outlying data and assigns most weight to more stable points.

Cubic spline curves are effective at interpolating between each consecutive data point, but can lose overall accuracy if the data is noisy due to outliers.
```python
def leastSquares(x,y):
    n = len(x)
    sumXi = np.sum(x)
    sumYi = np.sum(y)
    sumXiSquared = np.sum(x**2)
    sumXiYi = np.sum(x*y)
    
    m = (n*sumXiYi - sumXi*sumYi)/(n*sumXiSquared - (sumXi**2))
    c = (sumYi - m*sumXi)/n
    
    return m, c
```

```python
# Excluded Covid
x_no_covid = np.concatenate((x[:13], x[17:]))
y_no_covid = np.concatenate((y[:13], y[17:]))

# Least Squares Model Solve
m, c = leastSquares(x_no_covid, y_no_covid)
y_linear = m*x_no_covid + c

plt.scatter(x_no_covid, y_no_covid, color = 'blue', label = 'Given Data points')
plt.plot(x_no_covid, y_linear, color = 'red', label = 'Least Squares Curve')
plt.xlabel('Quarter')
plt.ylabel('GDP Growth (%)')
plt.title('U.S Quarterly GDP Growth Rate from 2010 to 2023 (excluding COVID Quarters)')
plt.legend()
plt.show()
print(f'Slope: {m}')

# Residual plot
residuals = y_no_covid - y_linear
plt.scatter(x_no_covid, residuals, color = 'blue', label = 'Residuals')
plt.xlabel('Quarter')
plt.ylabel('Residual')
plt.title('Residual Plot')
plt.legend()
plt.show()

# Root Mean Square Error
rmse = np.sqrt(np.mean((y_no_covid - y_linear)**2))
print("RMSE: ", rmse)
```

2. The slope of the least squares curve is approximately -.0019, which is very close to 0. This suggests that GDP growth does not show a significant long-term linear increase or decrease over time. A linear trend  does not make sense economically because it assumes GDP growth rate changes at a constant rate over time. In reality, GDP growth is affected by many factors, such as business cycles, changes in policy, and market shocks, causing it to vary irregularly.

PART C:

1. For a policymaker estimating GDP growth between two known points, a cubic spline is a better method for interpolation. Cubic splines ensure continuity in the first and second derivatives by using piecewise cubic functions, producing a smooth and realistic interpolation between data points. Unlike polynomial fitting, which shows an overall trend and is highly sensitive to outliers, cubic splines are locally defined, meaning that extreme values affect only surrounding segments rather than the entire curve. This difference in numerical stability is reflected in the condition numbers, where the tridiagonal spline system has a condition number of approximately 9, while the Vandermonde matrix for the polynomial fit is on the order of 10^8. Therefore, for interpolation between points, a cubic spline is smoother, more accurate, and more stable than a polynomial fit.
