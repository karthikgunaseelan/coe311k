```python
import numpy as np
import matplotlib.pyplot as plt

# Function representing dy/dx = -2y

def f(x, y):
    return -2 * y

def euler(step):
    
    # Define parameters
    
    h = step  # step size
    x_start, x_end = 0, 2
    num_steps = int((x_end - x_start) / h)

    x_values = np.linspace(x_start, x_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)

    # Initial condition
    
    y_values[0] = 1  # y(0) = 1
    
    # Implement Euler's method
    
    for i in range(num_steps):
        
        y_values[i + 1] = y_values[i] + h * f(x_values[i], y_values[i])
        
    return x_values, y_values
```

```python
plt.figure(figsize=(12, 8))
xref, yref = euler(0.01)

for i in np.arange(.01,1,.05):
    
    x_values,y_values = euler(i)
    
    current_yref = np.interp(x_values, xref, yref)
    pointwise_error = y_values - current_yref
    total_error = np.sum(np.abs(pointwise_error))
    
    print("Step size (h) = ", i)
    for x, err in zip(x_values, pointwise_error):
         print(f"{x:.4f}\t{err:.8f}")
    print(f"Total accumulated error: {total_error:.8f}")
    
    plt.plot(x_values, y_values, label=f"Step size = {i:.2f}")
    
# Plot the results

plt.xlabel('x')
plt.ylabel('y')
plt.title("Approximation of dy/dx = -2y using Euler's Method")
plt.legend()
plt.show()
```
