```python
import numpy as np
import time

def jacobi(A, b, x0, tolerance, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tolerance:
            return x_new, k
        x = x_new
    return x, max_iterations
```

```python
def gauss_seidel(A, b, x0, tolerance, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tolerance:
            return x_new, k
        x = x_new
    return x, max_iterations
```

```python
def solve(A, b, x0, tolerance, max_iterations):
    start = time.time()
    solution_jacobi, iterations_jacobi = jacobi(A, b, x0, tolerance, max_iterations)
    end = time.time()
    print(f"Jacobi Solution: {solution_jacobi}")
    print(f"Jacobi Iterations: {iterations_jacobi}")
    print(f"Elapsed Time: {end - start} sec\n")
    if iterations_jacobi == max_iterations:
        print("Jacobi Method diverges.\n")
    
    start = time.time()
    solution_gauss, iterations_gauss = gauss_seidel(A, b, x0, tolerance, max_iterations)
    end = time.time()
    print(f"Gauss-Seidel Solution: {solution_gauss}")
    print(f"Gauss-Seidel Iterations: {iterations_gauss}")
    print(f"Elapsed Time: {end - start} sec\n")
    if iterations_gauss == max_iterations:
        print("Gauss-Seidel method diverges.\n")
```

```python
A = np.array([
[0.582745, 0.48 , 0.10 , 0.  , 0.  ],
[0.48 , 1.044129, 0.46 , 0.10 , 0.  ],
[0.10 , 0.46 , 1.10431 , 0.44 , 0.10 ],
[0.  , 0.10 , 0.44 , 0.963889, 0.42 ],
[0.  , 0.  , 0.10 , 0.42 , 0.522565]], dtype=float)
b = np.array([1.162745, 2.084129, 2.20431 , 1.923889, 1.042565], dtype=float)
x0 = np.zeros_like(b)
tolerance = 1e-6
max_iterations = 100

solve(A, b, x0, tolerance, max_iterations)
```

```python
A = np.array([[1, 1], [1, 1.0001]])
b = np.array([1, 1])
x0 = np.zeros_like(b)
tolerance = 1e-6
max_iterations = 100

solve(A, b, x0, tolerance, max_iterations)
```

```python
A = np.array([[1, 1], [1, 1.0001]])
b = np.array([1, 1.0001])
x0 = np.zeros_like(b)
tolerance = 1e-6
max_iterations = 100

solve(A, b, x0, tolerance, max_iterations)
```
