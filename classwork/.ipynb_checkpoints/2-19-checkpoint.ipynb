Jacobi Method:
- Use 0 for x1,x2,x3 when solving for each, get answers
- Use those answers as new x1,x2,x3, compare with previous x1,x2,x3, and want to reach certain error or epsilon (extremly low error)

```python
import numpy as np
def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations
```

```python
# Example usage
A = np.array([[4, -1, 0, 0],
[-1, 4, -1, 0],
[0, -1, 4, -1],
[0, 0, -1, 3]])
b = np.array([15, 10, 10, 10])
x0 = np.zeros_like(b)
tol = 1e-6
max_iterations = 100
solution, iterations = jacobi(A, b, x0, tol, max_iterations)
print(f"Solution: {solution}")
print(f"Iterations: {iterations}")
```

Gauss-Seidel Method:
- Use 0 to find x1, then use that x1 value to find x2 (with x3=0 for now), then use those to find x3, and iterate that way
    - Basically same a Jacobi, but for every x you solve for, you plug that in instead of 0

Both Methods:
- If the matrices are not diagonally dominant, methods may not work as well

Error Analysis - Looking at each individual error between iterations
Convergence - How fast we get our answer
Tolerance - How close our error should be to epsilon
Residual Error - How well solution satisfies original equation
Covergence Speed - How fast the error decreases as iterations progress
