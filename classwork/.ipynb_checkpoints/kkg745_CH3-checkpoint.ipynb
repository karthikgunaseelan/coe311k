```python
def MonteCarlo(n):
    
    
    x = np.random.rand(n)
    y = np.random.rand(n)
    z = np.sqrt(x**2+y**2)
    
    z = z[z<1]
    
    estPi = 4*len(z)/n
    return estPi
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt

trials = np.arange(100)
trueError = np.zeros(100)
relativeError = np.zeros(100)

for i in range (0,100):
    xapproximate = MonteCarlo(1000)
    trueError[i] = abs(xapproximate - np.pi)
    relativeError[i] = abs((np.pi-xapproximate)/np.pi)
    
plt.plot(trials, trueError, 'b.')
plt.plot(trials, relativeError, 'r.')
```
