```python
import math

def calc_hyp(a,b):
    h = math.sqrt(a*a, b*b)
    return h
```

```python
import random

def approximatePi(numSamples):
    counter = 0
    for s in range(0,numSamples):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if (x*x + y*y < 1):
            counter = counter + 1
    return 4*counter/numSamples
```

```python
approximatePi(1000000)
```
