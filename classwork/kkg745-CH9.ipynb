```python
import numpy as np
import matplotlib.pyplot as plt

# Data points
X = np.array([1, 2, 3, 4, 5])
Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])
# Stack the design matrix with a column of ones (for the intercept)
A = np.vstack([X, np.ones(len(X))]).T
# Solve using least squares
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
print(f"Slope: {m}, Intercept: {c}")

plt.plot(X, Y, 'b.', label="Data Points")
plt.plot(X, m*X + c, 'r-', label="Least Squares Best-Fit")
plt.legend()
plt.show()
```

```python
X = np.random.rand(5)*10
X = np.sort(X)
Y = np.random.rand(5)*10
A = np.vstack([X, np.ones(len(X))]).T
# Solve using least squares
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
print(f"Slope: {m}, Intercept: {c}")

plt.plot(X, Y, 'b.', label="Data Points")
plt.plot(X, m*X + c, 'r-', label="Least Squares Best-Fit")
plt.legend()
plt.show()
```

Since X and Y were randomized independently, there is no true linear relationship between them. The least-squares line still exists, but it only minimizes error and does not represent a meaningful trend, so the fit may look poor or nearly flat.

```python
X = np.arange(0,1,0.01)
Y = 2*X + 1
error = np.random.rand(len(X))*0.5-.25
y = Y + error

A = np.vstack([X, np.ones(len(X))]).T
# Solve using least squares
m, c = np.linalg.lstsq(A, y, rcond=None)[0]

residuals = y - (m*X + c)
max_index = np.argmax(np.abs(residuals))
min_index = np.argmin(np.abs(residuals))
    
print(f"Slope: {m}, Intercept: {c}")
print(f"Residuals:\n{residuals}")
print(f"Maximum Residual: {residuals[max_index]} at ({X[max_index]},{y[max_index]})")
print(f"Minimum Residual: {residuals[min_index]} at ({X[min_index]},{y[min_index]})")

plt.plot(X,Y,'r-', label = 'Original Equation')
plt.plot(X, y, 'b.', label="Data Points")
plt.plot(X, m*X + c, 'g-', label = 'Best Fit Line')
plt.legend()
plt.show()
```
