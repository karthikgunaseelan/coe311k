```python
# Coefficient matrix A and constant vector b
A = np.array([[2, 3], [4, -1]])
b = np.array([5, 3])

# Solve the system A * x = b
x = np.linalg.solve(A, b)

print("Solution (x, y):", x)
```

Eigenvalue / Eigenvector : Provided there is a matrix a, there exists a scalar lambda and vector v such that the matrix times the vector = the scalar times that vector
- Eigenvectors - directions that dont change direction when a matrix acts on them
- Eigenvalues - tell you how much those directions get stretched or squished
   ● If eigenvalues are negative → system decays
   ● Positive → system grows
   ● Complex → oscillations

```python
A = np.array([[2,1],[1,2]])
eigenvalues, eigenvectors = np.linalg.eig(A)

print(eigenvalues)
print(eigenvectors)
```

```python
# Define a matrix
A = np.array([[4, 7], [2, 6]])

# Compute the inverse of A
A_inv = np.linalg.inv(A)

print("Inverse of A:\n", A_inv)

# Verify: A * A_inv should be the identity matrix (like [[1,0][0,1]])
identity = np.dot(A, A_inv)
print("A * A_inv:\n", identity)
```

Singular Value Decomposition (SVD) is a factorization of a matrix into three
matrices:
A=UΣV T
Where:
● U and V are orthogonal matrices,
● Σ is a diagonal matrix of singular values.

```python
# Define a matrix
A = np.array([[1, 2], [3, 4], [5, 6]])
# Perform SVD
U, sigma, Vt = np.linalg.svd(A)
print("U matrix:\n", U)
print("Sigma values:", sigma)
print("V^T matrix:\n", Vt)
```
