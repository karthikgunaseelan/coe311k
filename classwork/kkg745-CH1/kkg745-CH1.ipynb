```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,np.pi,100)
plt.subplot(2,2,1)
plt.plot(x,np.sin(x),2,'r')
plt.subplot(2,2,2)
plt.scatter(x,np.cos(x),2,'b')
plt.subplot(2,2,3)
plt.scatter(x,np.tan(x),2,'g')
plt.show()
plt.subplot(2,2,3)
plt.plot(x,np.arctan(x),2,'r')
```

```python


def f1(x,y):
    return (x*x+y*y)
x = np.linspace(-10, 10, 2000)
y = np.linspace(-10, 10, 2000)
X, Y = np.meshgrid(x, y)
Z = f1(X, Y)
```

```python
fig1 = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='hot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.view_init(10, 0)
```
