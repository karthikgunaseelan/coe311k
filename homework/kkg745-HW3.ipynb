```python
def sinTaylorSeries(x,n):
    sum = 0
    for i in range(n):
        sum = sum + (-1)**(i)*(x**(1+2*i)) / factorial(1+2*i) 
    return sum
```

```python
def factorial(n):
    sum = 1
    for i in range(1,n+1):
        sum = sum*i
    return sum
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

sinValues = np.zeros(101)

oneTermApprox = np.zeros(101)
twoTermApprox = np.zeros(101)
threeTermApprox = np.zeros(101)
fourTermApprox = np.zeros(101)

for i in range(0,101):
    x = .1*i
    sinValues[i] = np.sin(x)

for i in range(0,101):
    x = .1*i
    oneTermApprox[i] = sinTaylorSeries(x,1)

for i in range(0,101):
    x = .1*i
    twoTermApprox[i] = sinTaylorSeries(x,2)

for i in range(0,101):
    x = .1*i
    threeTermApprox[i] = sinTaylorSeries(x,3)

for i in range(0,101):
    x = .1*i
    fourTermApprox[i] = sinTaylorSeries(x,4)

oneTermError = np.abs(oneTermApprox - sinValues)
twoTermError = np.abs(twoTermApprox - sinValues)
threeTermError = np.abs(threeTermApprox - sinValues)
fourTermError = np.abs(fourTermApprox - sinValues)
```

```python
x = np.arange(0,10.1,.1)
plt.plot(x,oneTermError, "r", label="1 term")
plt.plot(x,twoTermError, "b", label="2 terms")
plt.plot(x,threeTermError, "g", label="3 terms")
plt.plot(x,fourTermError, "y", label="4 terms")

plt.xlabel("x-values")
plt.ylabel("Absolute Error")
plt.title("Taylor Series Error for sin(x)")
plt.legend()
plt.show()
```

As you increase the number of terms initially, specifically from 1 to 4 terms, the amount of error can increase for larger values of x. 

This is because the Taylor Series' higher power values of x grow quickly when there are less terms included in the sum, as the alternating effect is lessened.

However, when there are a large amount of terms, error decreases as the alternating effect of the series is more balanced.

As a result, error does not decrease uniformly for all values of x, as it changes based on term count and how large x is.
