```python
import numpy as np
import matplotlib.pyplot as plt

def leastSquares(x,y):
    n = len(x)
    xsum = np.sum(x)
    ysum = np.sum(y)
    xsumSquared = np.sum(x**2)
    xysum = np.sum(x*y)
    
    m = (n*xysum - xsum*ysum)/(n*xsumSquared - (xsum**2))
    c = (ysum - m*xsum)/n
    
    return m, c
```

```python
def residuals(X, y, m, c):
    r = y - (m*X + c)
    max_index = np.argmax(np.abs(r))
    min_index = np.argmin(np.abs(r))
    
    print(f"Slope: {m}, Intercept: {c}")
    print(f"Residuals:\n{r}")
    print(f"Maximum Residual: {r[max_index]} at ({X[max_index]},{y[max_index]})")
    print(f"Minimum Residual: {r[min_index]} at ({X[min_index]},{y[min_index]})")
```

```python
X = np.arange(0,1,0.01)
Y = 2*X + 1
error = np.random.rand(len(X))
y = Y + error

A = np.vstack([X, np.ones(len(X))]).T
# Solve using least squares
mNP, cNP = np.linalg.lstsq(A, y, rcond=None)[0]
mCustom, cCustom = leastSquares(X,y)

plt.plot(X,Y,'r-', label = 'Original Equation')
plt.plot(X, y, 'b.', label="Data Points")
plt.plot(X, mNP*X + cNP, 'g-', label = 'NumPy Least Square Method')
plt.legend()
plt.show()
residuals(X,y,mNP,cNP)

plt.plot(X,Y,'r-', label = 'Original Equation')
plt.plot(X, y, 'b.', label="Data Points")
plt.plot(X, mCustom*X + cCustom, 'g-', label = 'Custom Least Square Method')
plt.legend()
plt.show()
residuals(X,y,mCustom,cCustom)
```

The custom least squares method is very similar to identical to the original numpy least squares method.
