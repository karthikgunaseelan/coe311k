```python
def MonteCarlo(n):
    
    
    x = np.random.rand(n)
    y = np.random.rand(n)
    z = np.sqrt(x**2+y**2)
    
    z = z[z<1]
    
    estPi = 4*len(z)/n
    return estPi
    
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt

samplecount = np.arange(10,10000,20)
error = np.zeros(len(samplecount))

for i in range(0,len(samplecount)):
    n = samplecount[i]
    estPi = MonteCarlo(n)
    error[i] = abs(np.pi - estPi)
    
plt.plot(samplecount,error, 'm.')
plt.xlabel("Sample Count")
plt.ylabel("Error")
plt.title("Error Vs. Sample Count")
plt.show()
```
